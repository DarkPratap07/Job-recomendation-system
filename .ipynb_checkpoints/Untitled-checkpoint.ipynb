```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("data_nlp.csv", low_memory=False)

```

```python
df.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Count number of jobs per career level
career_counts = df['career level'].value_counts()

# Plot
plt.figure(figsize=(10,6))
sns.barplot(x=career_counts.index, y=career_counts.values, palette="viridis")

plt.title("Number of Jobs by Career Level", fontsize=16)
plt.xlabel("Career Level", fontsize=12)
plt.ylabel("Number of Jobs", fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
